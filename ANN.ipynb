# Regression

```python
# Importing the required libraries

import numpy as np
import pandas as pd
```

```python
# Reading the dataset

df = pd.read_csv('/content/50_Startups.csv')
```

```python
# Visualize the data

df.head()
```

```python
# Checking null values

df.isnull().sum()
```

```python
# Ckecking data type

df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Initializing encoder

le = LabelEncoder()
```

```python
# Conerting string type into int type

df['State'] = le.fit_transform(df['State'])
```

```python
df.head()
```

```python
# Split the data (Independent and dependent)

x = df.iloc[:,0:4].values
y = df.iloc[:,4:5].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Split the data (Training & testing)

xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
# Finding shapes od data

xtrain.shape, xtest.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Creating ANN skleton view

regressor = Sequential()
regressor.add(Dense(4,activation='relu'))
regressor.add(Dense(12,activation='relu'))
regressor.add(Dense(8,activation='relu'))
regressor.add(Dense(9,activation='relu'))
regressor.add(Dense(1,activation='linear'))
```

```python
# Compiling the ANN model

regressor.compile(optimizer='adam',loss='mse',metrics=['mse'])
```

```python
# Training the model

regressor.fit(xtrain,ytrain,batch_size=10,epochs=300)
```

```python
# Testing the model

ypred = regressor.predict(xtest)
```

```python
from sklearn.metrics import r2_score
```

```python
# Finding accuracy

r2_score(ytest,ypred)*100
```

```python
# converting 2d into 1d

ypred.flatten()
```

```python
# Comparing actual data with predicted data

pd.DataFrame({'Actual value':ytest.flatten(),
              'Predicted value':ypred.flatten()}).head(10)
```

```python
# Testing model with random data

regressor.predict([[1000,2354,1234,2]])
```

```python
# Testing model with random data

regressor.predict([[1000456,23235554,1232334,2]])
```

# Classification

```python
# Reading the dataset

data = pd.read_csv('/content/Breast Cancer Wisconsin (Diagnostic) Data Set.csv')
```

```python
# Visualize the data

data.head()
```

```python
# Checking null values

data.isnull().sum()
```

```python
# Remove unwanted columns

data.drop(['id','Unnamed: 32'],axis=1,inplace=True)
```

```python
# Ckecking data type

data.info()
```

```python
# Split the data (Independent and dependent)

x = data.drop('diagnosis',axis=1)
y = data['diagnosis']
```

```python
# Initializing encoder

le1 = LabelEncoder()
```

```python
# Converting string to int

y = le1.fit_transform(y)
```

```python
# Split the data (Training & testing)

xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
xtrain.shape
```

```python
# Creating ANN skleton view

classification = Sequential()
classification.add(Dense(30,activation='relu'))
classification.add(Dense(128,activation='relu'))
classification.add(Dense(64,activation='relu'))
classification.add(Dense(32,activation='relu'))
classification.add(Dense(1,activation='sigmoid'))
```

```python
# Compiling the ANN model

classification.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
# Training the model

classification.fit(xtrain,ytrain,batch_size=10,epochs=300)
```

```python
# Testing the model

ypred = classification.predict(xtest)
```

```python
# Comparing actual data with predicted data

pd.DataFrame({'Actual value':ytest.flatten(),
              'Predicted value':np.round(ypred.flatten()).astype('int')}).head(20)
```
